```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
a=[0,3,3,6,8,10,14,16,16,19,24]
np.quantile(a,0.25)
np.quantile(a,0.5)
np.quantile(a,0.75)
q1=np.quantile(a,0.25)
q2=np.quantile(a,0.5)
q3=np.quantile(a,0.75)
iqr=q3-q1
print(iqr)
plt.boxplot(a)
plt.text(1.05,q1,f"Q1:{q1:.2f}",color="r")
plt.text(1.05,q2,f"Q2:{q2:.2f}",color="r")
plt.text(1.05,q3,f"Q3:{q3:.2f}",color="r")
plt.show()
sns.distplot(a)
plt.show()
```

```python
def find_longest_word(words_list):
  word_len = []
  for n in words_list:
    word_len.append((len(n), n))
    word_len.sort()
  return word_len[-1][0], word_len[-1][1]

word_list = list(input().split(' '))
print(find_longest_word(word_list))
```

```python
class vehicle:
  def __init__(self,make,model,year):
    self.make=make
    self.model=model
    self.year=year
  def start_engine(self):
    print('staring engine')

class car(vehicle):
  def __init__(self,make,model,year,num_doors):
    super().__init__(make,model,year)
    self.num_doors=num_doors
  def drive(self):
    print('driving the car')
# now we r creating objects
a1=vehicle('toyoto','breeza',2021)
a2=car('suzuki','shift',2022,4)
a1.start_engine()
a2.drive()
a2.start_engine()
```
